1. 數據分析中時常會分析不同族群的資料 groupby()的使用
   分析男生與女生的各科平均值

分解groupby邏輯過程

Split：將大的數據集拆成可獨立計算的小數據集(拆成男生、女生資料)
Apply：獨立計算各個小數據集(成績取平均)
Combine：將小數據集運算結果合併

```python
import pandas as pf

result = pd.DataFrame([[1,50,80,70, 'F'],
                         [2,60,45,50,'F'],
                         [3,98,43,55,'F'],
                         [4,70,69,89,'F'],
                         [5,56,79,60,'M'],
                         [6,60,68,55,'M'],
                         [7,45,70,77,'M'],
                         [8,55,77,76,'M'],
                         [9,25,57,69,'M'],
                         [10,88,40,43,'M']],
                         columns = ['student_id','math_score', 'english_scrore', 'chinese_score', 'sex'])
result
```

```python
result.groupby('sex').mean()
```

2. Groupby 多個欄位做分析 (groupby 自動會生成多維度索引(multiple index))

group by class, sex，計算各科目的中位數

```python
#先增加class列
result['class'] = [1,2,1,2,1,2,1,2,1,2]
result
```

```python
result.groupby(['sex', 'class']).mean()
```

3. Groupby 欄位做多個分析 agg()

```python
result.groupby('sex').agg(['mean', 'std'])
```

4. Groupby 多個欄位做多個分析

```python
result.groupby(['sex', 'class']).agg(['mean', 'max'])
```

```python
score_df = pd.DataFrame([[1,50,80,70,'boy',1],
                         [2,60,45,50,'boy',2],
                         [3,98,43,55,'boy',1],
                         [4,70,69,89,'boy',2],
                         [5,56,79,60,'girl',1],
                         [6,60,68,55,'girl',2],
                         [7,45,70,77,'girl',1],
                         [8,55,77,76,'girl',2],
                         [9,25,57,60,'girl',1],
                         [10,88,40,43,'girl',3],
                         [11,25,60,45,'boy',3],
                         [12,80,60,23,'boy',3],
                         [13,20,90,66,'girl',3],
                         [14,50,50,50,'girl',3],
                         [15,89,67,77,'girl',3]],
                        columns=['student_id','math_score','english_score','chinese_score','sex','class'])
score_df
```

1. 找出全年級各科成績最高分與最低分？

```python
score_df.max(), score_df.min()
```
