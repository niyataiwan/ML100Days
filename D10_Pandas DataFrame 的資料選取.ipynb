## Study Summary

1. 用欄位名稱選取單行/多行資料
2. 用索引位置選取單列/多列資料
3. 用loc, iloc, iat, at選取行與列資料
4. 用遮罩選取資料

```python
import numpy as np
import pandas as pd
```

### 用欄位名稱選取單行資料

```python
x = pd.DataFrame([[1,2,3],[4,5,6]], index = ['a','b'], columns = ['A','B','C'])
print(x[['A','B','C']])
print('-----------')
print(x['B'])
```

### 用索引位置選取單列/多列資料

```python
Y = pd.DataFrame([[1,2,3],['niya','tina','gina']], index = ['no.','name'], columns = ['A','B','C'])
print(Y)
print('------------------')
print(Y[0:1])
```

### 用loc, iloc, iat, at選取行與列資料

```python
#loc: 用index, columns選取資料
Z = pd.DataFrame([[1,2,3],['niya','tina','gina']], index = ['no.','name'], columns = ['A','B','C'])
print(Z)
print('----------------------')
print(Z.loc['no.','C'])
print('----------------------')
print(Z.loc['no.',['A','B']])
print('----------------------')
print(Z.loc[['no.','name'],['A','C']])
```

```python
#iloc: 用指標選取資料
i = pd.DataFrame([[1,2,3],['niya','tina','gina']], index = ['no.','name'], columns = ['A','B','C'])
print(i)
print('----------------------')
print(i.iloc[0,2])
print('----------------------')
print(i.iloc[[0],[0,1]])
print('----------------------')
print(i.iloc[[0,1],[0,2]])
print('----------------------')
print(i.iloc[[0,1],[0,2]])
```

### 用遮罩選取資料

```python
u = pd.DataFrame([[1,2,3],['niya','tina','gina']], index = ['no.','name'], columns = ['A','B','C'])
print(u.loc['no.'] > 1)
```

### 2. 請根據提供的資料，選擇出下列的要求
       select the first 3 rows
       select the odd rows(index =1,3,5)
       select the last 2 columns
       select the even columns(index = 0,2,4)

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(np.random.randn(6,3), columns=['A','B','C'])
print(df)
```

```python
#select the first 3 rows
df[0:3]
```

```python
#select the odd rows(index =1,3,5)
df.loc[1::2]
```

```python
#select the last 2 columns
df.iloc[-2:]
```

```python
#select the even columns(index = 0,2,4)
print(df.iloc[0::2])
print('---------------------------')
print(df.loc[0::2])
```

### 3. 請根據提供的資料，選擇出下列的要求
       filtered by first column > 20
       filtered by first column + second column  > 50
       filtered by first column < 30 or second column > 30
       filtered by total sum of row > 100

```python
u = pd.DataFrame(np.random.randint(10,40,60). reshape(-1,4))
u
```

```python
#filtered by first column > 20
u[u.iloc[:,0] > 20]   #印出全部DataFrame,並篩掉第一列>20的資料
```

```python
#filtered by first column + second column  > 50
u[u.iloc[:,0] + u.iloc[:,1] > 50]
```

```python
#filtered by first column < 30 or second column > 30
u[(u.iloc[:,0] < 30) | (u.iloc[:,1]>30)]
```

```python
#filtered by total sum of row > 100
u[u.sum(axis = 1) > 100]
```
